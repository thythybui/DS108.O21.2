c) Cấu trúc các tập tin của bộ dữ liệu? Ý nghĩa từng tập tin.


Cấu trúc thư mục gồm các tập tin chính sau:

README.txt: Giới thiệu tổng quát bộ dữ liệu.

features_info.txt: Mô tả chi tiết về các biến đặc trưng (features).

features.txt: Danh sách tên của tất cả 561 đặc trưng.

activity_labels.txt: Ánh xạ giữa nhãn (số) và tên hoạt động (vd: 1 → WALKING).

Thư mục train/ và test/:

X_train.txt / X_test.txt: Ma trận đặc trưng của tập huấn luyện / kiểm tra (mỗi hàng tương ứng 1 mẫu với 561 đặc trưng).

y_train.txt / y_test.txt: Nhãn tương ứng cho mỗi mẫu trong X_train / X_test.

subject_train.txt / subject_test.txt: ID người thực hiện hoạt động trong từng mẫu.

Thư mục con Inertial Signals/:
Lưu tín hiệu thô theo từng trục (X, Y, Z) cho mỗi loại dữ liệu:

total_acc_[x|y|z]_train.txt: Gia tốc toàn phần.

body_acc_[x|y|z]_train.txt: Gia tốc phần thân (sau khi loại bỏ trọng lực).

body_gyro_[x|y|z]_train.txt: Vận tốc góc (từ con quay hồi chuyển).

3) Merge bộ dữ liệu train và test lại với nhau.

Merge dữ liệu:
- Merge dữ liệu train: merge column giữa X_train, y_train và subject_train. Đặt tên biến là X_train_data.
	Sử dụng hàm cbind theo thứ tự: (y_train, subject_train, X_train). 
	Đổi tên cột y_train thành "Activity", cột subject_train thành "Subject". Sử dụng hàm names(X_train_data)[1] <- "Activity ".
- Merge dữ liệu test: merge column giữa X_test, y_test và subject_test. Đặt tên biến là X_test_data
	Thực hiện tương tự như merge dữ liệu train.


```python
X_train_data = pd.concat([Y_train, subject_train, X_train], axis=1)
```

```python
print(X_train_data.shape)
print(X_train_data.head())
```

```python
X_train_data.columns = ['Activity', 'Subject'] + list(X_train_data.columns[2:])
X_train_data.head()
```

```python
X_test_data = pd.concat([Y_test, subject_test, X_test], axis=1)
print(X_test_data.shape)
```

```python
X_test_data.columns = ['Activity', 'Subject'] + list(X_test_data.columns[2:])
X_test_data.head()
```

- Merge dữ liệu giữa tập train và test lại với nhau.
	Sử dụng hàm rbind - gộp dòng. Gộp dữ liệu giữa 2 bộ X_train_data và X_test_data lại với nhau.


```python
X_data = pd.concat([X_train_data, X_test_data], axis=0)
X_data
```
